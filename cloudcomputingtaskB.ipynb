```python
import pandas as pd
import itertools 
```

```python
pd.read_csv('AComp_Passenger_data_no_error(1).csv')
```

```python
pd.read_csv('Top30_airports_LatLong(1).csv')
```

```python
passenger_data = []
with open('AComp_Passenger_data_no_error(1).csv', 'r') as f:
    for line in f:
        fields = line.strip().split(',')
        passenger_id = fields[0]
        flight_count = 1
        passenger_data.append((passenger_id, flight_count))
```

```python
def mapper(passenger_tuple):
    passenger_id, flight_count = passenger_tuple
    return (passenger_id, flight_count)

def reducer(passenger_id, flight_counts):
    return (passenger_id, sum(flight_counts))
```

```python
# Map phase
mapped_passengers = map(mapper, passenger_data)

# Shuffle and sort phase
sorted_passengers = sorted(mapped_passengers, key=lambda x: x[0])

# Reduce phase
reduced_passengers = []
for passenger_id, flights in itertools.groupby(sorted_passengers, key=lambda x: x[0]):
    flight_counts = map(lambda x: x[1], flights)
    reduced_passengers.append(reducer(passenger_id, flight_counts))

# Find the passenger(s) with the highest number of flights
max_flights = max(reduced_passengers, key=lambda x: x[1])[1]
top_passengers = filter(lambda x: x[1] == max_flights, reduced_passengers)

# Output the results
for passenger in top_passengers:
    print("Passenger {} has taken {} flights.".format(passenger[0], passenger[1]))
```
